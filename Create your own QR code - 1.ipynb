```python
# Import qrcode and make from link
import qrcode 
img = qrcode.make("https://www.geeksforgeeks.org/python-pil-image-show-method/") # Enter your link
img.save("myQRcode.png")
# img.show()
```

```python
#Show the QR code
from PIL import Image
import matplotlib.pyplot as plt
im=Image.open("myQRcode.png")
plt.figure(figsize=(5,6))
plt.imshow(im)
plt.axis("off")
plt.grid(False)
plt.show()
```
